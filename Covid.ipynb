```python
#- bibliotecas

import pandas as pd
import numpy as py
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/MarcusAbrantes/Documents/HIST_PAINEL_COVIDBR_2020-23.csv', sep=';')
df
```

```python
#Renomeei algumas colunas que iremos usar para nossa análise
df = df.rename(columns={'regiao': 'Região', 'estado': 'Estado', 'data': 'Data', 'casosAcumulado': 'Casos Acumulado', 'casosNovos': 'Novos Casos', 'obitosAcumulado': 'Óbitos Acumulados', 'obitosNovos': 'Novos Óbitos', 'coduf': 'Cód. UF', 'semanaEpi': 'Semana Epidemológica', 'populacaoTCU2019': 'População em 2019'})
df
```

```python
#Retirei colunas que não achei necessárias
df=df.drop(['codmun', 'codRegiaoSaude', 'nomeRegiaoSaude', 'Recuperadosnovos', 'emAcompanhamentoNovos', 'interior/metropolitana', 'municipio'], axis=1)
df
```

```python
#Transformei a data para datetime
df['Data'] = pd.to_datetime(df['Data'])
df
```

```python
# Criei uma coluna de semestres com base na coluna Data
df['Semestre'] = df['Data'].dt.year.astype(str) + '-' + ((df['Data'].dt.month - 1) // 6 + 1).astype(str)
df
```

```python
# Adicionei colunas para facilitar os óbitos e os casos
df['Milhões de Casos'] = df['Novos Casos'] / 1000000
df['Milhares de Óbitos'] = df['Novos Óbitos'] / 1000
df
```

```python
#Gráfico de calor representando novos caso (em milhões) por semestre e região

df_pivot_regiao_semestre = df.pivot_table(index='Região', columns='Semestre', values='Milhões de Casos', aggfunc='sum')
plt.figure(figsize=(10, 6))
sns.heatmap(df_pivot_regiao_semestre, cmap='YlOrBr', annot=True, fmt='g', cbar=True)
plt.xlabel('Semestre')
plt.ylabel('Região')
plt.title('Distribuição dos Novos Casos por Semestre e por Região')
plt.show()
```

```python
#Gráfico de calor representando novos óbitos (em milhares) por semestre e região

df_pivot_regiao_semestre = df.pivot_table(index='Região', columns='Semestre', values='Milhares de Óbitos', aggfunc='sum')
plt.figure(figsize=(10, 6))
sns.heatmap(df_pivot_regiao_semestre, cmap='coolwarm', annot=True, fmt='g', cbar=True)
plt.xlabel('Semestre')
plt.ylabel('Região')
plt.title('Distribuição dos Novos Óbitos por Semestre e por Região')
plt.show()
```

```python
#Gráfico para calcular o total de mortes por estado

total_mortes_estado = df.groupby('Estado')['Óbitos Acumulados'].max()

total_mortes_estado = total_mortes_estado.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=total_mortes_estado.index, y=total_mortes_estado.values)
plt.xlabel('Estado')
plt.ylabel('Total de Mortes')
plt.title('Total de Mortes por Estado durante a Pandemia')
plt.xticks(rotation=90)
plt.show()
```

```python
#Gráfico que mostra a proporção de casos acumulados em relação à população de cada estado

df['Proporção Casos População'] = df['Casos Acumulado'] / df['População em 2019']
df_estado_proporcao_casos = df.groupby('Estado')['Proporção Casos População'].max()
print(df_estado_proporcao_casos)
df_estado_proporcao_casos.plot(kind='bar', xlabel='Estado', ylabel='Proporção Casos População', title='Proporção de Casos Acumulados em Relação à População por Estado')
plt.show()
```

```python
#Gráfico que mostra o total de casos acumulados por estado

df_estado_obitos = df.groupby('Estado')['Novos Casos'].sum()
print(df_estado_obitos)
df_estado_obitos.plot(kind='bar', xlabel='Estado', ylabel='Novos Casos', title='Total de Casos Acumulados por Estado')
plt.show()
```

```python
# Gráfico para mostrar a porcentagem da população que pegou morreu de Covid em cada Estado
df_periodo = df[(df['Data'].dt.year >= 2020) & (df['Data'].dt.year <= 2023)]

df_por_estado = df_periodo.groupby('Estado').agg({'Novos Óbitos': 'sum', 'População em 2019': 'mean'}).reset_index()

df_por_estado['Porcentagem da População Perdida'] = (df_por_estado['Novos Óbitos'] / df_por_estado['População em 2019']) * 100

df_por_estado = df_por_estado.sort_values('Porcentagem da População Perdida', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=df_por_estado, x='Estado', y='Porcentagem da População Perdida', order=df_por_estado['Estado'])
plt.xlabel('Estado')
plt.ylabel('Porcentagem da População Perdida')
plt.title('Porcentagem da População Perdida por Estado')
plt.xticks(rotation=90)
plt.show()
```
